# <b> Problem 8 : Create a new column for the updated price. This updated price will be calculated on top of the given price - if the engine is in front, price will be same else if the engine is in rear, price will be doubled </b>


```python
# Your code here. Use apply and lambda

automobile_df['new_prices'] = automobile_df.apply(lambda x: x['price'] if x['engine-location'] == 'front' else x['price']*2, axis=1)
automobile_df.tail(10)
```

```python
# Another way to do same task
def calculate_new_price(row):
    if row['engine-location'] == 'front':
        return row['price']
    else:
        return row['price'] * 2
```

```python
automobile_df["new_price"] = automobile_df.apply(calculate_new_price, axis = 1)
automobile_df.head(10)
```

# <b> Problem 10 : Create a new column which stores the number of doors in a car as integers.

```python
def number_of_door(door):
  if door == 'two':
    return 2
  elif door == "four" :
    return 4
  else:
    return 0
```

# <b>Problem 11 : Concatenate the two data frames given below firstly row wise and secondly column wise.</b>

<code>GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}</code>

<code>japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}</code>

```python
# Your code here

# Creating germancars DataFrames from the given dictionaries
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMW', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
germancars_df = pd.DataFrame(GermanCars)
germancars_df
```

```python
# Creating japanesecars DataFrames from the given dictionaries
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi'], 'Price': [29995, 23600, 61500 , 58900]}
japanesecars_df = pd.DataFrame(japaneseCars)
japanesecars_df
```

```python
# Concatenate row-wise
car_df_row_concat = pd.concat([germancars_df , japanesecars_df], axis=0, ignore_index=True)

car_df_row_concat
```

```python
# Concatenate column-wise
car_df_col_concat = pd.concat([germancars_df , japanesecars_df], axis=1)

car_df_col_concat
```
