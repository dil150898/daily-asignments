# <b><u>Problem 2 </u></b>

### Read face cream and facewash product sales data and show it using the bar chart. The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```
# Comparing face wash and face cream number of unit shold every month

face_cream = sales_df["facecream"]
fash_wash = sales_df["facewash"]
month_number = sales_df["month_number"]
x_axis = np.arange(len(month_number))
plt.bar(x_axis - 0.2, face_cream, 0.38, label = 'Face cream' , color = "green")
plt.bar(x_axis + 0.2, fash_wash, 0.38, label = 'Face wash' , color = "cyan")
plt.xticks(x_axis, month_number)
plt.xlabel("Month Number")
plt.ylabel("Number of units sold")
plt.title("Number of units sold per month for each product")
plt.legend()
plt.show()
```

# <b><u>Problem 3</u></b>

### Display the number of units sold per month for each product using multiline plots. (i.e. separate plotline for each product ).

```
# Reading number of unit for each items sold every month

months = sales_df['month_number']
facecream = sales_df['facecream']
facewash = sales_df['facewash']
toothpaste = sales_df['toothpaste']
bathingsoap = sales_df['bathingsoap']
shampoo = sales_df['shampoo']
moisturizer = sales_df['moisturizer']

# Plotting each product's sales
plt.plot(months, facecream, marker='|', label='Face Cream', linestyle='-', linewidth=1)
# As fashwash and moisturizer sold eqally every month i have increased linewidth of fash was that it can be visualize in graph
plt.plot(months, facewash, marker='|', label='Face Wash', linestyle='-', linewidth=2 , color = 'cyan')
plt.plot(months, toothpaste, marker='|', label='Toothpaste', linestyle='-', linewidth=1)
plt.plot(months, bathingsoap, marker='|', label='Bathing Soap', linestyle='-', linewidth=1)
plt.plot(months, shampoo, marker='|', label='Shampoo', linestyle='-', linewidth=1)
plt.plot(months, moisturizer, marker='|', label='Moisturizer', linestyle='-', linewidth=1 , color = 'black')

# Adding titles and labels
plt.title('Number of Units Sold Per Month for Each Product', fontsize=16, fontweight='bold')
plt.xlabel('Month Number', fontsize=14)
plt.ylabel('Units Sold', fontsize=14)

plt.legend()

# Adding grid
plt.grid(True)

# Display the plot
plt.show()


```
